Projeto de Machine Learning focado na classificação de clientes inadimplentes em uma empresa de empréstimos de automóveis. Neste trabalho, são aplicadas diferentes técnicas de otimização de hiperparâmetros para melhorar o desempenho dos modelos, analisando como ajustes nos hiperparâmetros impactam os resultados e a capacidade de previsão dos algoritmos.
